## Use Basement data to predict

```python
# load Library

import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt

from sklearn import preprocessing


%matplotlib inline
```

```python
Bsm_Sales_DF = pd.read_csv('C:Data/Bsm_Sales_DF.csv',
                               sep=',',
                               header=0,
                               index_col=0,
                               )
```

```python
import xgboost

model = xgboost.XGBRegressor(colsample_bytree=0.4,
                 gamma=0,                 
                 learning_rate=0.07,
                 max_depth=3,
                 min_child_weight=1.5,
                 n_estimators=10000,                                                                    
                 reg_alpha=0.75,
                 reg_lambda=0.45,
                 subsample=0.6,
                 seed=42) 

model.fit(Bsm_Sales_DF.dropna().drop('SalePrice', axis=1),
          Bsm_Sales_DF.dropna().loc[:,'SalePrice'])

model.score(Bsm_Sales_DF.dropna().drop('SalePrice', axis=1),
          Bsm_Sales_DF.dropna().loc[:,'SalePrice'])
```

```python
from xgboost import plot_importance
# plot feature importance
plot_importance(model)
```

### Basic Linear Regression

```python
# First do a Linear Regression 

from sklearn.linear_model import LinearRegression
reg = LinearRegression().fit(Bsm_Sales_DF.dropna().drop('SalePrice', axis=1),
                             Bsm_Sales_DF.dropna().loc[:,'SalePrice'])

# score of the model
reg.score(Bsm_Sales_DF.dropna().drop('SalePrice', axis=1),
          Bsm_Sales_DF.dropna().loc[:,'SalePrice'])

```

#### Quantile Transformation

```python
# transform the distribution to analyse with a correlation matrix
quantile_transformer = preprocessing.QuantileTransformer(output_distribution='normal',
                                                         random_state=0)

# Trasform the DF to a array 
Lot_Array = Bsm_Sales_DF.dropna().drop('SalePrice', axis=1).values

# quatile trasnformation of the Lot values
Lot_Array = quantile_transformer.fit_transform(Lot_Array)

#Transform the Array in a DF
Lot_DF = pd.DataFrame(Lot_Array,
                      columns=Bsm_Sales_DF.dropna().drop('SalePrice', axis=1).columns)

# Linear Regression after quantiletrasnformer

reg_quant = LinearRegression().fit(Lot_DF,
                             Bsm_Sales_DF.dropna().loc[:,'SalePrice'])

# score of the model
reg_quant.score(Lot_DF,
                Bsm_Sales_DF.dropna().loc[:,'SalePrice'])

```

#### yeo-johnson transformation

```python
# transform the distribution to analyse with a correlation matrix
pt = preprocessing.PowerTransformer(method='yeo-johnson',
                                    standardize=True)
# Trasform the DF to a array 
Lot_Array = Bsm_Sales_DF.dropna().drop('SalePrice', axis=1).values

# quatile trasnformation of the Lot values
Lot_Array = pt.fit_transform(Lot_Array)

#Transform the Array in a DF
Lot_DF = pd.DataFrame(Lot_Array,
                      columns=Bsm_Sales_DF.dropna().drop('SalePrice', axis=1).columns)

# Linear Regression after quantiletrasnformer

reg_yeo = LinearRegression().fit(Lot_DF,
                             Bsm_Sales_DF.dropna().loc[:,'SalePrice'])

# score of the model
reg_yeo.score(Lot_DF,
              Bsm_Sales_DF.dropna().loc[:,'SalePrice'])

```

### Ridge regression

### ridge with quant transformation

```python
# transform the distribution to analyse with a correlation matrix
quantile_transformer = preprocessing.QuantileTransformer(output_distribution='normal',
                                                         random_state=0)

# Trasform the DF to a array 
Lot_Array = Bsm_Sales_DF.dropna().drop('SalePrice', axis=1).values

# quatile trasnformation of the Lot values
Lot_Array = quantile_transformer.fit_transform(Lot_Array)

#Transform the Array in a DF
Lot_DF = pd.DataFrame(Lot_Array,
                      columns=Bsm_Sales_DF.dropna().drop('SalePrice', axis=1).columns)

# Ridge CV 
ridgeCV_quant = RidgeCV(alphas=[1e-3, 1e-2, 1e-1, 1, 1e+2],
                        gcv_mode='auto',
                        normalize=False,
                        scoring='r2').fit(Bsm_Sales_DF.dropna().drop('SalePrice', axis=1),
                                                    Bsm_Sales_DF.dropna().loc[:,'SalePrice'])

# score of the model
ridgeCV_quant.score(Bsm_Sales_DF.dropna().drop('SalePrice', axis=1),
                    Bsm_Sales_DF.dropna().loc[:,'SalePrice'])
```
